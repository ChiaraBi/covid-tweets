## This notebook merge all the csv files in the "data" folder

```python
#libraries

import os
import glob
import pandas as pd
```

```python

os.chdir("../data")
file_extension = '.csv'
all_filenames = [i for i in glob.glob(f"*{file_extension}")]
print(f"These are all the .csv files that will be merged: {all_filenames}.")
```

### merge all the csv

```python
all_tweets = pd.concat([pd.read_csv(f) for f in all_filenames if not os.stat(f).st_size == 0])
all_tweets =pd.DataFrame(all_tweets)

# remove duplicates
all_tweets.drop_duplicates(subset='id',inplace=True)

all_tweets
```

### select only useful columns
Change cell below or don't execute it if you need different columns

```python
all_tweets=all_tweets[['id_str','text']]
all_tweets
```

### Export as csv 
Note: the file is in the "src" folder in order to avoid to mess up with the files in the "data" folder

```python
os.chdir("../src")
all_tweets.to_csv('all_tweets.csv',index=False)
```
